# 01. OpenCV Webcam Basics

**Goal:** Create a "Hello World" camera app that runs in a loop.


## 1. The Computer Vision Loop
Videos are just fast sequences of images (Frames). We need a `while` loop that:
1.  **Reads** a frame from the camera.
2.  **Processes** it (draws stuff).
3.  **Displays** it.
4.  **Waits** for a quit key.


## 2. Opening the Camera


```python
import cv2

# '0' is usually the default webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open camera.")
else:
    print("Camera detected!")
    cap.release()
```

## 3. The Main Loop (Run this!)


```python
import cv2
import time

cap = cv2.VideoCapture(0)

print("Press 'q' to quit the window.")

while True:
    # 1. Read
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break
        
    # 2. Process (Draw a rectangle)
    # cv2.rectangle(img, start_point, end_point, color_bgr, thickness)
    cv2.rectangle(frame, (50, 50), (250, 250), (255, 0, 0), 2)
    
    # Add Text
    cv2.putText(frame, "Hello OpenCV", (50, 40), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # 3. Display
    cv2.imshow('Lesson 01', frame)
    
    # 4. Wait (1ms) and check for 'q' key
    # 0xFF is a hex mask to get the last 8 bits
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print("Closed.")
```

### 🛠️ Engineer Thinking
- **Latency**: Everything inside the `while` loop adds delay. Keep it fast.
- **BGR Colors**: OpenCV uses Blue-Green-Red, not RGB. `(255, 0, 0)` is Blue, not Red.


```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow("hi", frame)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
