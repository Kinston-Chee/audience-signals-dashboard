# 07. Time Series Smoothing

**Goal:** Fix the "Flickering" problem.


## 1. The Problem
In one frame, yaw might be 0.64 (Front). Next frame: 0.66 (Away). The signal toggles rapidly. This is annoying.


## 2. The Solution: Moving Average
We keep a history of the last 30 frames (approx 1 second). We take the average.
- If 29 frames say "Front" and 1 says "Away", the average is still "Front".


```python
import collections

# Create a buffer of fixed length
history = collections.deque(maxlen=10)

# Simulate Data
import random
for i in range(20):
    val = 1 if i < 15 else 0 # Sudden drop
    history.append(val)
    
    # Calculate Smooth Average
    avg = sum(history) / len(history)
    print(f"Raw: {val} | Smooth: {avg:.2f}")
```

### 🛠️ Engineer Thinking
Larger window = Smoother data, but **Higher Latency**. 30 frames is a good sweet spot.
