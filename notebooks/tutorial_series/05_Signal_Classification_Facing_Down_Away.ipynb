# 05. Signal Classification

**Goal:** Turn math ratios into discrete Logic Signals (True/False).


## 1. Defining States
We define 3 core states:
1.  **Face Front**: Yaw and Pitch are both "Centered".
2.  **Eyes Front**: Iris is centered.
3.  **Eyes Open**: Not blinking.


## 2. Thresholds
- Yaw: 0.35 to 0.65
- Pitch: 0.30 to 0.70
- Iris: 0.35 to 0.65
- EAR (Eye Aspect Ratio): > 0.20 (Open)


```python
def get_signals(yaw, pitch, ear, iris):
    # 1. Face Logic
    is_face_front = False
    if yaw > 0.35 and yaw < 0.65 and pitch > 0.30 and pitch < 0.70:
        is_face_front = True
    
    # 2. Eye Logic
    is_eye_open = False
    if ear > 0.20:
        is_eye_open = True
        
    is_eye_front = False
    if is_eye_open:
        if iris > 0.35 and iris < 0.65:
            is_eye_front = True
    
    return {
        "Face Front": is_face_front,
        "Eyes Front": is_eye_front,
        "Eyes Open": is_eye_open
    }
```

### 🛠️ Engineer Thinking
Why separate Face and Eyes? Because you can face the screen but look away (Side-eye)!


```python
# Run this to test the logic with random numbers
print(get_signals(0.5, 0.5, 0.3, 0.5)) # Should be All True
print(get_signals(0.9, 0.5, 0.3, 0.5)) # Face Front False (Yaw too high)
print(get_signals(0.5, 0.5, 0.1, 0.5)) # Eyes Open False (Blinking)
```
