## 1. What We Are Building
We are building a **Real-time Audience Engagement Settings Dashboard** using your webcam.

### ⚠️ The Golden Rule (Ethics)
We use Computer Vision to detect **Physical Signals** (Head direction, Eye openness), NOT internal states (Boredom, Attention).

- **OK:** "The head is turned 40 degrees left."
- **NOT OK:** "The user is not paying attention."

We aggregate data (e.g., "30% of the audience is looking away") to protect individual privacy.


## 2. Tools We Use
1.  **OpenCV**: For reading the webcam and drawing boxes (The "Eyes").
2.  **MediaPipe**: For detecting Face Landmarks (The "Brain").
3.  **NumPy**: For high-speed math (The "Calculator").
