# 11. AGM Demo Script & Reliability

**Goal:** Presenting your work to stakeholders.


## 1. The Script (1 Minute)
> "Good morning. This tool measures aggregate audience engagement signals."
> "It uses computer vision to detect head direction and eye openness locally on the device."
> "It does not identify individuals."
> "Here you see the metrics: 80% of the audience is facing the screen."


## 2. Reliability Checklist
- [ ] Clean Camera Lens.
- [ ] Good Lighting (Face detected?).
- [ ] Close other Camera Apps (Teams/Zoom).
- [ ] Restart Kernel before demo starts.
