# 03. Landmarks & Coordinates

**Goal:** Understand the 468-point Face Mesh map.


## 1. Normalized Coordinates
MediaPipe returns `x, y, z` values between **0.0 and 1.0**.
- `x=0`: Left edge.
- `x=0.5`: Center.
- `x=1.0`: Right edge.

To draw them, we multiply by image width/height.


## 2. Key Landmarks map
We don't need all 468 points. We mostly care about:
- **Nose Tip**: Index `1`
- **Chin**: Index `152`
- **Left Eye Outer**: Index `33`
- **Right Eye Outer**: Index `263`
- **Left Iris Center**: Index `468`
- **Right Iris Center**: Index `473`


```python
import cv2
import mediapipe as mp

# Setup
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(max_num_faces=1, refine_landmarks=True)
cap = cv2.VideoCapture(0)

print("Showing Key Landmarks... 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret: break
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb)
    h, w, c = frame.shape
    
    if results.multi_face_landmarks:
        for face in results.multi_face_landmarks:
            
            # Helper to draw a point
            def draw_pt(idx, color, label):
                pt = face.landmark[idx]
                px, py = int(pt.x * w), int(pt.y * h)
                cv2.circle(frame, (px, py), 4, color, -1)
                cv2.putText(frame, label, (px+5, py), cv2.FONT_HERSHEY_PLAIN, 1, color, 1)
            
            # Draw main components
            draw_pt(1, (0, 255, 0), "Nose")
            draw_pt(152, (0, 255, 0), "Chin")
            draw_pt(33, (255, 0, 0), "L_Eye")
            draw_pt(263, (255, 0, 0), "R_Eye")
            draw_pt(468, (0, 255, 255), "L_Iris")

    cv2.imshow('Lesson 03', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): break

cap.release()
cv2.destroyAllWindows()
```
