# 08. The Dashboard UI

**Goal:** Create a professional overlay.


## 1. Transparency (Alpha Blending)
We draw a semi-transparent black box behind the text to make it readable on any background.


```python
import cv2
import numpy as np

def draw_panel(image, x, y, w, h, alpha=0.6):
    overlay = image.copy()
    cv2.rectangle(overlay, (x, y), (x+w, y+h), (0,0,0), -1)
    return cv2.addWeighted(overlay, alpha, image, 1 - alpha, 0)
```

## 2. Drawing Bars
A "Health Bar" for attention.


```python
def draw_bar(image, x, y, pct, color):
    width = int(pct * 2) # Scale 100% to 200px
    cv2.rectangle(image, (x, y), (x+width, y+10), color, -1)
    cv2.rectangle(image, (x, y), (x+200, y+10), (100,100,100), 1)

# Try it on a dummy image
img = np.zeros((300, 400, 3), dtype=np.uint8)
img = draw_panel(img, 10, 10, 250, 150)
draw_bar(img, 20, 50, 80, (0,255,0)) # 80%
cv2.putText(img, "Attention: 80%", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)
```
